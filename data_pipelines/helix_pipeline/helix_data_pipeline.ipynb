```python
sql_stage = sql_connector('DNAPROD', 'Stage')
```

```python
source = sql_stage.read_query('select top(10) CreatedDtm, Vendor_Id, FirstName, SurName from Stage.temp.PRTL_Vendor')
```

```python
source['Vendor_Id'] = source.Vendor_Id.apply(lambda x: int(x))
source['CreatedDtm'] = source.CreatedDtm.apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S') if x != '' else np.nan)
```

```python
target = source.drop(0).reset_index(drop = True)
source.loc[1, 'SurName'] = 'Bulka'
```

```python
# find all rows in the source table which are not present in the target table
changes = source[~source.apply(tuple, 1).isin(target.apply(tuple, 1))]
changes
```

```python
# check which rows in the source were modified
modified = changes[changes.Vendor_Id.isin(target.Vendor_Id)]
modified
```

```python
# check which rows in the source were added
added = changes[~changes.Vendor_Id.isin(target.Vendor_Id)]
added
```
